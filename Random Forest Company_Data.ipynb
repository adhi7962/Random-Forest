```python
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
```

```python
data= read_csv('D:\Assignment\Random Forest\Company_Data.csv')
```

```python
data.head()
```

```python
#Complete company dataset
label_encoder= preprocessing.LabelEncoder()
data['ShelveLoc']= label_encoder.fit_transform(data['ShelveLoc']) 
```

```python
data['Urban'].replace(to_replace=['Yes', 'No'],value= ['0', '1'], inplace=True)
data['US'].replace(to_replace=['Yes', 'No'],value= ['0', '1'], inplace=True)
data
```

```python
x=data.drop(columns=['ShelveLoc'])
y=data['ShelveLoc']
```

```python
y
```

```python
x
```

```python
num_trees = 100
max_features = 3
kfold = KFold(n_splits=10)
```

```python
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, x, y, cv=kfold)
print(results.mean())
```
